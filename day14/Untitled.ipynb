```python
import requests
from bs4 import BeautifulSoup

url='http://search.chosun.com/search/news.search'
query='코로나'
params = {
    'query':query, #검색할 키워드
    'pageno':1, #페이지 번호
    # 'orderby':'docdatetime', #정렬 순서(최신순)
}

page=requests.get(url,params)

page_soup = BeautifulSoup(page.content,'html.parser')

a_tags=page_soup.select('dt>a')
```

```python
news=list()
```

```python
a_tags
```

```python
for a in a_tags:
    print(a.get_text(),a['href'],sep='\n')
    news.append([a['href'],a.get_text()])
```

```python
for i in range(len(news)):
    inner_page=requests.get(news[i][0])
    inner_pate_soup=BeautifulSoup(inner_page.content,'html.parser')
    contents=inner_pate_soup.find('div',{'id':'news_body_id'})
    news[i].append(contents.get_text().replace('\n','').replace('\t','').strip())
```

```python
for new in news:
    for line in new:
        print(line)
    print('---------------------------'*4)
```
