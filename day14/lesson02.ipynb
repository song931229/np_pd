```python
import requests
from bs4 import BeautifulSoup
```

```python
url='https://movie.naver.com/movie/sdb/rank/rmovie.nhn'

page=requests.get(url)
soup=BeautifulSoup(page.text,'html.parser')
```

```python
soup
```

```python
targets=soup.find_all('div',class_='tit3')
```

```python
targets
```

```python
for target in targets:
    print(target.get_text().strip())
```

```python
targets2=soup.select('td> div > a')
```

```python
targets2
```

```python
for target2 in targets2:
    print(target2['title'])
```
