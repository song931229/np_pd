```python
p = re.compile('text: \"(.*?)\"', re.DOTALL)
```

```python
p.findall(text)
```

```python
target_index=-1
for i,x in enumerate(p.findall(text)):
    if "주식의 총수" in x:
        target_index=i
target_index
```

```python
re.findall('viewDoc\((.*)\)',text)
```

```python
target_text=re.findall('viewDoc\((.*)\)',text)[target_index]
target_text
```

```python
querys=target_text.replace("'","").replace(' ','').split(',')
querys
```

```python
url3='http://dart.fss.or.kr/report/viewer.do'
```

```python
params3={
    'rcpNo': querys[0],
    'dcmNo': querys[1],
    'eleId': querys[2],
    'offset': querys[3],
    'length': querys[4],
    'dtd': querys[5],
    'crtfc_key': api_key
}
```

```python
resq3=requests.get(url=url3, params=params3, headers=session.headers, timeout=120)
```

```python
data2=BeautifulSoup(resq3.content,'html.parser')
data2
```

```python
target_table=data2.find('table',{'border':1})
```

```python
target_th=None
target_tr=None
for x in target_table.find_all('tr'):
    if str(x).find('유통주식수')>=0:
        target_tr=x
    if str(x).find('보통주')>=0:
        target_th=x
print(target_th)
print(target_tr)
```

```python
for i,z in enumerate(target_th.find_all('td')):
    if '보통주' in z.text:
        target_index2=i
```

```python
stock_mounts=[]
for y in target_tr.find_all('td'):
    try:
        stock_mounts.append(int(y.text.replace(',','')))
    except:
        print(y.text)
```

```python
print('보통주(유통주식중)=',stock_mounts[target_index2])
```
