```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

# 글꼴 설정
font_path = "C:\Windows\Fonts\H2HDRM.TTF"
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
```

```python
df_crime=pd.read_csv('crime_in_Seoul.csv')
```

```python
df_crime.head()
```

```python
df_crime=df_crime.sort_values('구별')
df_crime.head()
```

```python
df_crime=df_crime.drop('Unnamed: 0',axis=1)
```

```python
pivot_crime=pd.pivot_table(data=df_crime,index='구별')
```

```python
pivot_crime
```

```python
pivot_crime.iloc[:,0::2].head()
```

```python
pivot_crime['검거 합계']=pivot_crime.iloc[:,0::2].sum(axis=1)
```

```python
pivot_crime.iloc[:,1::2].head()
```

```python
pivot_crime['발생 합계']=pivot_crime.iloc[:,1::2].sum(axis=1)
```

```python
pivot_crime.head()
```

```python
pivot_crime['검거율']=(pivot_crime['검거 합계']/pivot_crime['발생 합계'])*100
```

```python
pivot_crime.sort_values('검거율',inplace=True)
```

```python
pivot_crime.index
```

```python
# 범죄합계와 검거합계, 검거율을 그래프로 나타내기
flg = plt.figure(figsize=(16, 10))
ax1 = flg.add_subplot(1, 1, 1)
plt.subplots_adjust(hspace=0.3)
ax1.bar([x*2 for x in range(25)], pivot_crime['발생 합계'])
ax1.bar([x*2+1 for x in range(25)], pivot_crime['검거 합계'],tick_label=pivot_crime.index)
ax2 = ax1.twinx()
ax2.plot([x*2+0.5 for x in range(25)], pivot_crime['검거율'], marker='o', color='r')

# print(pivot_crime['발생 합계'], pivot_crime['검거 합계'], pivot_crime['발생 합계'] - pivot_crime['발생 합계'])
ax1.set_ylabel("발생수")
ax2.set_ylabel("검거율")
print(ax2.get_xticklabels())
ax1.set_xticklabels(ax1.get_xticklabels(),rotation=45)

plt.show()
```

```python
# 각 범죄에 대한 검거율을 따로 구했다
pivot_crime['강간검거율'] = pivot_crime['강간 검거']/pivot_crime['강간 발생']*100
pivot_crime['강도검거율'] = pivot_crime['강도 검거']/pivot_crime['강도 발생']*100
pivot_crime['살인검거율'] = pivot_crime['살인 검거']/pivot_crime['살인 발생']*100
pivot_crime['절도검거율'] = pivot_crime['절도 검거']/pivot_crime['절도 발생']*100
pivot_crime['폭력검거율'] = pivot_crime['폭력 검거']/pivot_crime['폭력 발생']*100

# 검거율이 100이 넘는 곳이 나와서 100이 넘으면 100으로 만들어 주자
con_list = ['강간검거율', '강도검거율', '살인검거율', '절도검거율', '폭력검거율']
for column in con_list:
      pivot_crime.loc[pivot_crime[column] > 100, column] = 100

fig = plt.figure(figsize=(20, 20))
plt.subplots_adjust(wspace=0.2, hspace=0.4)
crime = pivot_crime.loc[:, ['강간검거율', '강도검거율', '살인검거율', '절도검거율', '폭력검거율']]

label = ['강간', '강도', '살인', '절도', '폭력']
explode = [0, 0, 0, 0.2, 0]

for i in range(25):
      ax = fig.add_subplot(5, 5, i+1)
      ax.pie(crime.iloc[i], labels=label, explode=explode, startangle=40)
      ax.set_title(crime.index[i])
plt.show()
```
