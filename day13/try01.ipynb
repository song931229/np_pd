```python
import urllib
```

```python
searchlist=['파이썬','웹 크롤링','빅데이터','python']
```

```python
url='www.naver.com'
```

```python
for x in searchlist:
    url='http://search.naver.com/search.naver?query='+urllib.parse.quote(x)
    resp=urllib.request.urlopen(url)
    byte_data=resp.read()
    text_data=byte_data.decode()
    print(text_data)
```

```python
parsed=urllib.parse.urlsplit(url)
```

```python
parsed=list(parsed)
```

```python
parsed[3]=urllib.parse.quote('호랑이')
```

```python
unparsed=urllib.parse.urlunsplit(parsed)
```

```python
unparsed
```

```python
resp=urllib.request.urlopen(unparsed)
byte_data=resp.read()
text_data=byte_data.decode()
print(text_data)
```
