```python
import requests
```

```python
url='https://datalab.naver.com/'
page=requests.get(url)
temp=page.text.split('<span class="title_cell">2020.06.10.(수)</span>')[1]
```

```python
for x in temp.split('<span class="title">')[1:-5]:
    print(x.split('</span>')[0])
```
