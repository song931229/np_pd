# DATA불러와서 조작해보기

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# !move "C:\/Users/\i7B/\Downloads\/sampledata/\시도별 전출입 인구수.xlsx"
```

```python
df=pd.read_excel("시도별 전출입 인구수.xlsx",header=0)
df=df.fillna(method='ffill')
df
```

```python
mask=(df['전출지별']=='서울특별시')&(df['전입지별']!='서울특별시')
mask
```

```python
df2=df[mask]
df2
```

```python
df3=df2.drop(['전출지별'],axis=1)
df3
```

```python
df4=df3.rename({'전입지별':'전입지'},axis=1)
df4
```

```python
df5=df4.set_index('전입지')
df5
```

```python
dfa=df5.loc['경기도']
```

```python
dfa
```

## 그래프에 마커넣어보기
- 마커의 종류
https://matplotlib.org/api/markers_api.html

```python
plt.figure(figsize=(30,20))
plt.plot(dfa.index,dfa.values,marker='o',markersize=12)
import matplotlib.font_manager as fm
font_location='C:/Windows/Fonts/H2PORL.TTF'
font_name=fm.FontProperties(fname=font_location,size=18)
plt.title('서울->경기 이동인구',fontproperties=font_name)
plt.xlabel('기간',fontproperties=font_name)
plt.ylabel('이동인구수',fontproperties=font_name)

plt.show()
```

> ## 그래프 변경하기
- 막대그래프로 변경
- 전체 글꼴 적용
- xlabel 돌리기
- 여라가지 스타일 적용해보기

```python
print(plt.style.available)
```

```python
from matplotlib import font_manager,rc
font_path='C:/Windows/Fonts/H2PORL.TTF'
font_name=font_manager.FontProperties(fname=font_path).get_name()
rc('font',family=font_name)
# plt.style.use('Solarize_Light2')
plt.figure(figsize=(14,10))
plt.xticks(rotation='vertical')
plt.ylim(50000,700000)
plt.bar(dfa.index,dfa.values)
plt.title('서울->경기 이동인구',fontsize=25)
plt.xlabel('기간',fontsize=18)
plt.ylabel('이동인구수',fontsize=18)

plt.legend(('인구수'),fontsize=20,loc='upper center')

plt.show()
```

```python
plt.figure(figsize=(30,20))
plt.plot(dfa.index,dfa.values,marker='o',markersize=12)

plt.title('서울->경기 이동인구',fontsize=25)
plt.xlabel('기간',fontsize=18)
plt.ylabel('이동인구수',fontsize=18)

plt.ylim(50000,700000)

plt.annotate('',
             xy=(20, 620000),       # 화살표의 끝점
             xytext=(2, 290000),    # 화살표의 시작점
             xycoords='data',
             arrowprops=dict(arrowstyle='->', color='skyblue', lw=5)) # 화살표 서식

plt.annotate('인구이동증가(1970~1995)',  # 텍스트 내용
             xy=(10, 400000),           # 텍스트의 위치 기준점
             rotation=30,               # 텍스트의 회전 각도
             va='baseline',             # 텍스트의 상하 정렬
             ha='center',               # 텍스트의 좌우 정렬
             fontsize=15)               # 텍스트 크기

plt.annotate('',
             xy=(40, 450000),       # 화살표의 끝점
             xytext=(25, 620000),    # 화살표의 시작점
             xycoords='data',
             arrowprops=dict(arrowstyle='->', color='red', lw=5)) # 화살표 서식

plt.show()
```
