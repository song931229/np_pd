# 그래프 그리기

```python
#그래프그리는 라이브러리
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=np.linspace(0,2*np.pi)
y=np.sin(x)
x
y
```

## 그래프의 생성과 출력

```python
plt.plot(x,y)
plt.show()
```

## x,y축 표시 범위 지정

```python
plt.xlim([0,5])
plt.ylim([0,1])
plt.plot(x,y)
plt.show()
```

## 그래프에 제목 부여하기

### 그래프 제목 나타내기

```python
plt.title('title')
plt.show()
```

### X축 제목 나타내기

```python
plt.xlabel("x-axis")
plt.show()
```

### Y축 제목 나타내기

```python
plt.ylabel("y-axis")
plt.show()
```

### 축 제목 한글 넣기

```python
plt.xlabel("x축")
plt.ylabel("y축")
plt.show()
```

```python
import matplotlib.font_manager as fm
font_location='C:/Windows/Fonts/H2PORL.TTF'
font_name=fm.FontProperties(fname=font_location,size=18)
```

```python
plt.xlabel("x축",fontproperties=font_name)
plt.ylabel("y축",fontproperties=font_name)
plt.show()
```

## 그래프 눈금(grid) 표시하기
### 기본눈금 적용하기

```python
plt.grid(True)
plt.plot(x,y)
plt.show()
```

### 그래프 축에 눈금 설치 하기

```python
positions=[np.pi/2*x for x in range(5)]
labels=[str(90*x) for x in range(5)]
plt.xticks(positions, labels)
plt.grid(True)
plt.plot(x,y)
plt.show()
```

```python
x= np.linspace(0,2*np.pi)
y=np.sin(x)
z=np.cos(x)
plt.title("제목",fontproperties=font_name)
plt.xlabel("x축",fontproperties=font_name)
plt.ylabel("y축",fontproperties=font_name)
plt.grid(True)
plt.xticks(positions, labels)
plt.plot(x,y,color='r', label='y=sin(x)')
plt.plot(x,z,color='b', label='z=cos(x)')
plt.legend(["y=sin(x)","z=cos(x)"])
plt.show()
```

```python
x= np.linspace(0,2*np.pi)
y=np.sin(x)
z=np.cos(x)
t=np.tan(x)
plt.title("제목",fontproperties=font_name)
plt.xlabel("x축",fontproperties=font_name)
plt.ylabel("y축",fontproperties=font_name)
plt.plot(x,y,color='r', label='y=sin(x)')
plt.plot(x,z,color='b', label='z=cos(x)')
plt.plot(x,t,color='g', label='t=tan(x)')
```
