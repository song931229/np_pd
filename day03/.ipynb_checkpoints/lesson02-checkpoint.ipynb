## 서브플롯:그래프 여러개 그리기

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

- 서브플롯 단위로 그래프를 그리는 영역을 지정할 수 있다.
- 해당 영역마다 다른 그래프를 그릴 수 있다.

```python
x=np.linspace(0,2*np.pi)
y=np.sin(x)

#크기를지정해서 생성(단위 인치)
fig = plt.figure(figsize=(4,4))

ax=fig.add_subplot(2,3,5)

plt.grid(True)
plt.plot(x,y)
plt.show()
```

### 서브플롯 그려보기

```python
x=np.linspace(0,2*np.pi)
y=np.sin(x)
z=np.cos(x)
t=np.tan(x)

#크기를지정해서 생성(단위 인치)
fig = plt.figure(figsize=(9,6))

ax1=fig.add_subplot(2,3,5)
ax2=fig.add_subplot(2,3,4)
ax3=fig.add_subplot(2,3,3)

ax1.plot(x,y)
ax2.plot(x,z)
ax3.plot(x,t)
plt.ylim(-1,1)

for i in range(6):
    if i ==4or i==3 or i==2:
        continue
    fig.add_subplot(2,3,i+1)
plt.show()
```

### 서브플롯간의 간격 조절하기

```python
plt.subplots_adjust(wspace=1,hspace=2)
plt.show()
```

### 서브플롯의 그래프 표시 범위 설정

```python
ax3.set_ylim(-1,1)
```

### 서브플롯의 그래프 이름 설정

```python
ax1.set_title('y=sin(x)')
```

```python
plt.show()
```

### 서브플롯의 grid설정

```python
ax1.grid(True)
plt.show()
```

### 서브플롯의 라벨설정

```python
positions=[np.pi/2*x for x in range(5)]
labels=[str(90*x) for x in range(5)]
ax1.set_xticks(positions)
ax1.set_xticklabels(labels)
```

### 실습

```python
x1=[x for x in range(10)]
x2=np.linspace(-2*np.pi,2*np.pi)
display(x2)
f1=[x**1 for x in range(10)]
f2=[x**2 for x in range(10)]
f3=[x**3 for x in range(10)]

f4=np.sin(x2)
f5=np.cos(x2)
f6=np.tan(x2)

flist=[f1,f2,f3,f4,f5,f6]

fig = plt.figure(figsize=(9,6))

axL=[None]*6

titles=['x^1','x^2','x^3','sin(x)','cos(x)','tan(x)']


for i in range(3):
    axL[i]=fig.add_subplot(2,3,i+1)
    axL[i].plot(x1,flist[i])
    
for i in range(3):
    axL[i+3]=fig.add_subplot(2,3,i+4)
    axL[i+3].plot(x2,flist[i+3])

for i in range(3):
    axL[i+3].set_ylim(-1,1)

axL[3].set_xlim(np.pi*(-1),np.pi*(1))    
axL[4].set_xlim(np.pi*(-1),np.pi*(1))
axL[5].set_xlim(np.pi*(-0.5),np.pi*(0.5))
for i in range(6):
    axL[i].grid()
    axL[i].set_title(titles[i])
    
plt.subplots_adjust(wspace=0.4,hspace=0.4)
plt.show()
```
