```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from matplotlib import font_manager,rc
font_path='C:/Windows/Fonts/H2PORL.TTF'
font_name=font_manager.FontProperties(fname=font_path).get_name()
rc('font',family=font_name)
```

```python
df=pd.read_excel("시도별 전출입 인구수.xlsx",header=0)
df=df.fillna(method='ffill')
df.tail()
```

```python
mask1=(df['전출지별']=='제주특별자치도')&(df['전입지별']!='제주특별자치도')
mask2=(df['전출지별']!='제주특별자치도')&(df['전입지별']=='제주특별자치도')
df_a=df[mask1]
df_b=df[mask2]
df_a.head()
```

```python
df_b.head()
```

```python
df_a=df_a.drop(df.loc[:,'1970':'1999'],axis=1)
```

```python
df_b=df_b.drop(df.loc[:,'1970':'1999'],axis=1)
```

```python
df_a
```

```python
df_a=df_a.replace('-','0')
df_b=df_b.replace('-','0')
```

```python
df_a.loc[:,'2000':]=df_a.loc[:,'2000':].astype('int')
df_b.loc[:,'2000':]=df_b.loc[:,'2000':].astype('int')
```

```python
df_a.head()
```

```python
df_b.head()
```

```python
df_a.drop(307,inplace=True)
df_b.drop(18,inplace=True)
```

```python
df_a.head()
```

```python
move_out=df_a.loc[:,"2000":].sum()
```

```python
move_in=df_b.loc[:,"2000":].sum()
```

```python
move_out
```

```python
move_in
```

```python
move_in.index
```

```python
plt.figure(figsize=(15,10))
plt.bar([x*2 for x in range(18)],move_in)
plt.bar([x*2+1 for x in range(18)],move_out)
plt.xticks([x*2+0.5 for x in range(18)],move_in.index,rotation=45)
plt.title("00~17 제주도 전출입")
plt.xlabel('년도')
plt.ylabel('인구수')
plt.legend(['전입','전출'])
```
