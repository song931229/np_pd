```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('classic')

file_path = "auto-mpg.csv"
df = pd.read_csv(file_path, header=None)

df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight',
              'acceleration', 'model year', 'origin', 'name']

df
```

## 연비(mpg)열에 대한 히스토그램 그리기

```python
df['mpg'].plot(kind='hist', bins=10, figsize=(10, 5))

plt.show()
```

## 연비(mpg)와 차종(weight)에 대한 산점도 그리기

### 기본산점도

```python
cylinders_size=df.cylinders/df.cylinders.max()*360
```

```python
df.plot(kind='scatter', x='weight', y='mpg' ,c='coral')
plt.show()
```

### 산점도의 농도(크기)

```python
df.plot(kind='scatter', x='weight', y='mpg' ,c='coral', s=cylinders_size, figsize=(15,10))
plt.savefig('auto-mpg.png')
```
