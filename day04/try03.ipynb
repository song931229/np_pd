```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

font_path = "C:\Windows\Fonts\H2HDRM.TTF"
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

df=pd.read_excel('../day03/남북한발전전력량.xlsx')
df.fillna(method="ffill",inplace=True)
df
```

```python
df = df.drop(['전력량 (억㎾h)'], axis=1)
df.set_index('발전 전력별', inplace=True)

col_years = list(map(str, range(1990, 2017)))
df_hap = df.loc[['합계'], col_years]
hap = df_hap.transpose()
hap.columns = ['남한전력', '북한전력']
print(hap)

# 스타일 서식 지정
plt.style.use('ggplot')

hap.index = hap.index.map(int)

# bar : 세로, barh : 가로
hap.plot(kind='bar', figsize=(10, 7), width=0.7)

plt.title("남북한 발전전력량", size=30)
plt.ylabel('발전 전력량', size=20)
plt.xlabel('기간', size=20)

plt.show()
```
