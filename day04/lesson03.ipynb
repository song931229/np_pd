# arear그래프

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

# 글꼴 설정
font_path = "C:\Windows\Fonts\H2HDRM.TTF"
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

# 파일 가져오기
file_path = "../day03/시도별 전출입 인구수.xlsx"
df = pd.read_excel(file_path, header=0)
df = df.fillna(method='ffill')

# 파일에서 서울에서 전입지별만 뽑아오기
mask = (df['전출지별'] == '서울특별시') & (df['전입지별'] != '서울특별시')
df_seoul = df[mask]
df_seoul = df_seoul.drop(['전출지별'], axis=1)
df_seoul.rename({'전입지별':'전입지'}, axis=1, inplace=True)
df_seoul.set_index('전입지', inplace=True)

col_years = list(map(str, range(1970, 2018)))
df_4 = df_seoul.loc[['충청남도', '경상북도', '강원도', '전라남도'], col_years]
display(df_4)

df_4 = df_4.transpose()
df_4
```

```python
# 스타일 서식 지정
plt.style.use('ggplot')

# 인덱스의자료형바꾸기
df_4.index=df_4.index.map(int)
df_4
```

```python
df_4.plot(kind='area')
plt.show()
```

## 선명도줄이기

```python
df_4.plot(kind='area',alpha=0.2)
plt.show()
```

## 라벨링, 제목설정 etc

```python
#크기
df_4.plot(kind='area',alpha=0.2, figsize=(20,10))

#제목
plt.title('서울->타시도 이동 ', size=30)
#y라벨
plt.ylabel('이동 인구 수', size=20)
#x라벨
plt.xlabel('기간', size=20)
plt.show()
```

## stacked 옵션

```python
#크기
df_4.plot(kind='area',alpha=0.2, stacked=0, figsize=(5,5))
```

```python
#크기
df_4.plot(kind='area',alpha=0.2, stacked=1, figsize=(5,5))
```

```python
axe=df_4.plot(kind='area',alpha=0.2, stacked=1, figsize=(5,5))
#제목
axe.set_title('서울->타시도 이동 ', size=30)
#y라벨
axe.set_ylabel('이동 인구 수', size=20)
#x라벨
axe.set_xlabel('기간', size=20)
plt.show()
```
