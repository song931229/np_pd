# 꺾은선그래프
## 꺾은선 그래프의 다양한 속성

```python
import numpy as np
import matplotlib.pyplot as plt
days = np.arange(1,11)
weight=np.array([10,14,18,20,18,16,17,18,20,17])

plt.plot(days,weight,marker='D', markerfacecolor='b' , linestyle=':')
plt.show()
```

> maker의 종류  
- o = 원,
- s = 사각형
- p = 오각형
- * = 별
- + = 플러스
- D = 마름모  
...etc

> color의 종류 (markerface,color)
- b = blue
- r = red
- c = cyan
- m = magenta
- y = yellow
- k = black
- w = white  
...etc

>linestyle의 종류
- - = 실선
- -- = 파선
- -. = 1점쇄선
- : = 점선  
...etc

# 막대 그래프

```python
x = [x for x in range(1,7)]
y = [12,41,32,46,21,17]
y2=[43,1,6,17,17,9]

labels=['Apple','Orange','Banana','Pineapple','Kiwifruit','StrawBerry']

#라벨 적용
plt.bar(x,y, tick_label=labels)
#누적 막대 그래프
plt.bar(x,y2,bottom=y)
plt.show()
```

# 히스토그램

```python
np.random.seed(34)
data=np.random.randn(10000)

plt.show()
```

## 구간수 설정하기

```python
plt.hist(data, bins=100)#bins 구간수설정
plt.show()
```

```python
plt.hist(data, bins='auto')
plt.show()
```

## 정규화하기

```python
plt.hist(data, bins='auto', density=True)
plt.show()
```

## 누적시키기

```python
plt.hist(data, bins='auto', density=True, cumulative=True)
plt.show()
```

# 산포도(scatter)

```python
np.random.seed(34)
x = np.random.choice(np.arange(100), 100)
y = np.random.choice(np.arange(100), 100)
z = np.random.choice(np.arange(100), 100)
plt.scatter(x, y)
plt.show()
```

## 마커변경하기

```python
plt.scatter(x, y, marker="+")
plt.show()
```

## 색 변경하기

```python
plt.scatter(x, y, marker="+",color='r')
plt.show()
```

## 농도로 표현하기

```python
plt.scatter(x, y, marker="+", s=z)
plt.show()
```

## 농도에 따른 색부여하기

```python
plt.scatter(x, y, marker="+", c=z, cmap='Purples')
plt.show()
```

## 농도에 따른 색상표 부여하기

```python
plt.scatter(x, y, marker="+", c=z, cmap='Purples')
plt.colorbar()
plt.show()
```

# 원형 그래프

```python
data=[60,20,10,5,3,2]
plt.pie(data)
plt.axis('equal')#적어주면 완벽한원형
plt.show()
```

## 라벨링

```python
labels=['Apple','Orange','Banana','Pineapple','Kiwifruit','StrawBerry']
```

```python
plt.pie(data, labels=labels)
plt.show()
```

## 조각 떼어내기(explode 0~1)

```python
explode=[0,0,0.5,0,0,0]
plt.pie(data, labels=labels, explode=explode)
plt.show()
```
