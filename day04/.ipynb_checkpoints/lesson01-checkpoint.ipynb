```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

file_path = "../day03/시도별 전출입 인구수.xlsx"
df = pd.read_excel(file_path, header=0)
df = df.fillna(method='ffill')

mask = (df['전출지별'] == '서울특별시') & (df['전입지별'] != '서울특별시')
df_seoul = df[mask]
df_seoul = df_seoul.drop(['전출지별'], axis=1)
df_seoul.rename({'전입지별': '전입지'}, axis=1, inplace=True)
df_seoul.set_index('전입지', inplace=True)
df_one = df_seoul.loc['경기도']

font_path = "C:\Windows\Fonts\H2HDRM.TTF"
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

plt.style.use('ggplot')
```

```python
fig=plt.figure(figsize=(10,10))

ax1= fig.add_subplot(2,1,1)
ax2= fig.add_subplot(2,1,2)
#그래프의 종류를 바꿔 그리기
ax1.plot(df_one,'o',markersize=10)
#그래프에 마커를 추가하기
ax2.plot(df_one, marker='o', markerfacecolor='green', markersize=10,
        color='olive', linewidth=2, label='서울->경기')
#범례위치
ax2.legend(loc='best')
#범위지정
ax1.set_ylim(50000,800000)
ax2.set_ylim(50000,800000)

#라벨 설정하고, 기울이기
ax1.set_xticklabels(df_one.index, rotation=75)
ax2.set_xticklabels(df_one.index, rotation=75)

plt.show()
```

## 그래프 작성해보기
> 서울->강원  
서울->충남  
서울->경북  
서울->경기  


```python
df_a=df_seoul.loc['강원도']
df_b=df_seoul.loc['충청남도']
df_c=df_seoul.loc['경상북도']
df_d=df_seoul.loc['경기도']
```

```python
fig=plt.figure(figsize=(10,40))
ax_1= fig.add_subplot(4,1,1)
ax_2= fig.add_subplot(4,1,2)
ax_3= fig.add_subplot(4,1,3)
ax_4= fig.add_subplot(4,1,4)

ax_1.plot(df_a, marker='o', markerfacecolor='white', markersize=10, color='red', linewidth=2, label='서울->강원')
ax_2.plot(df_b, marker='o', markerfacecolor='white', markersize=10, color='orange', linewidth=2, label='서울->충남')
ax_3.plot(df_c, marker='o', markerfacecolor='white', markersize=10, color='yellow', linewidth=2, label='서울->경북')
ax_4.plot(df_d, marker='o', markerfacecolor='white', markersize=10, color='green', linewidth=2, label='서울->경기')

ax_1.legend(loc='best')
ax_2.legend(loc='best')
ax_3.legend(loc='best')
ax_4.legend(loc='best')

#범위지정
ax_1.set_ylim(0,60000)
ax_2.set_ylim(0,60000)
ax_3.set_ylim(0,60000)
ax_4.set_ylim(0,700000)

#라벨 설정하고, 기울이기
ax_1.set_xticklabels(df_one.index, rotation=90)
ax_2.set_xticklabels(df_one.index, rotation=90)
ax_3.set_xticklabels(df_one.index, rotation=90)
ax_4.set_xticklabels(df_one.index, rotation=90)

plt.show()
```

```python
con=['강원도','충청남도','경상북도','경기도']
colors=['purple','navy','blue','green']
df_L=[None]*4
ax_L=[None]*4

figg=plt.figure(figsize=(10,40))

for x in range(4):
    df_L[x]=df_seoul.loc[con[x]]
    ax_L[x]=figg.add_subplot(4,1,x+1)
    
for x in range(4):
    ax_L[x].plot(df_L[x],marker='o', markerfacecolor='white', markersize=10, color=colors[x], linewidth=2, label=str(x))
    ax_L[x].set_ylim(0,60000)
    ax_L[x].set_xticklabels(df_one.index, rotation=90)

ax_L[3].set_ylim(0,700000)

plt.show()
```
