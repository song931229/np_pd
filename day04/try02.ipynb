```python
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import font_manager, rc

font_path = "C:\Windows\Fonts\H2HDRM.TTF"
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

df=pd.read_excel('../day03/남북한발전전력량.xlsx')
df=df.fillna(method='ffill')
```

```python
df
```

```python
mask = (df['전력량 (억㎾h)'] == '남한')
df_south = df[mask]
df_south
```

```python
df_south = df_south.drop(['전력량 (억㎾h)'], axis=1)
df_south
```

```python
df_south.set_index('발전 전력별', inplace=True)
df_south
```

```python
df_one = df_south.loc['합계']
df_one
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(14, 7))
plt.xticks(rotation='vertical')

plt.title('남한 발전 전력량', fontsize=18)
plt.xlabel('기간', fontsize=12)
plt.ylabel('전력량', fontsize=12)

plt.plot(df_one, marker='o', markersize=10)

plt.legend(labels=['남한발전전력량'], loc='best', fontsize=15)
plt.show()
```

```python
mask2 = (df['전력량 (억㎾h)'] == '북한')
df_north = df[mask2]
df_north = df_north.drop(['전력량 (억㎾h)'], axis=1)
df_north.set_index('발전 전력별', inplace=True)
df_two = df_north.loc['합계']
df_two
```

```python
fig=plt.figure(figsize=(10,10))

ax1= fig.add_subplot(2,1,1)
ax2= fig.add_subplot(2,1,2)

ax1.plot(df_one, marker='o', color='blue',markersize=10, label='남한 발전량')
ax2.plot(df_two, marker='o', color='red', markersize=10, label='북한 발전량')

ax1.legend(loc='best')
ax2.legend(loc='best')

ax1.set_xticklabels(df_one.index, rotation=75)
ax2.set_xticklabels(df_one.index, rotation=75)

plt.show()
```

```python
dfn[dfn=='-']=0
dfs[dfs=='-']=0
```

```python
dfn.loc['원자력']
```

```python
con=['수력','화력','원자력','신재생']

dfnL=[None]*4
for x in range(3):
    dfnL[x]=dfn.loc[con[x]]
    
dfnL[3]=dfn.loc[con[2]]
```

```python
dfsL=[None]*4
for x in range(4):
    dfsL[x]=dfs.loc[con[x]]
```

```python
fig=plt.figure(figsize=(15,7))
axL=[None]*8
for x in range(8):
    axL[x]= fig.add_subplot(2,4,x+1)

for x in range(4):
    axL[x].plot(dfnL[x], marker='o', markerfacecolor='pink', markersize=10,
                color='red', linewidth=2, label='북한')
    
for x in range(4):
    axL[x+4].plot(dfsL[x], marker='o', markerfacecolor='skyblue', markersize=10,
                  color='blue', linewidth=2, label='남한')
    
for x in range(8):
    axL[x].set_xticklabels(df_one.index, rotation=90)
plt.show()
```

```python
fig=plt.figure(figsize=(15,10))

ax_a= fig.add_subplot(2,1,1)
ax_b= fig.add_subplot(2,1,2)
for x in range(4):
    ax_a.plot(dfnL[x], marker='o', markersize=5, label=con[x])
ax_a.legend(loc='best')

for x in range(4):
    ax_b.plot(dfsL[x], marker='o', markersize=5, label=con[x])
ax_b.legend(loc='best')
plt.show()
```
