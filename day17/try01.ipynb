```python
import requests
```

```python
from bs4 import BeautifulSoup as bs
```

```python
url='https://sports.news.naver.com/kbaseball/record/index.nhn'
```

```python
data={
    'category': 'kbo'
}
```

```python
resp=requests.get(url,params='data')
```

```python
soup=bs(resp.content,'html.parser')
```

```python
soup
```

```python
tbody=soup.find('tbody')
```

```python
thead=soup.find('thead')
```

```python
rows=tbody.find_all('tr')
```

```python
heads=thead.find_all('span')
```

```python
columns=[]
for head in heads:
    columns.append(head.get_text().strip())
```

```python
print(len(columns))
columns
```

```python
tbody
```

```python
rows
```

```python
data=[]
```

```python
for row in rows:
    data.append(row.text.split('\n'))
```

```python
data
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(data)
```

```python
df.drop([0,2,3,4,6,7,18],axis=1,inplace=True)
```
