# seaborn 활용하기
- seaborn : matplotlib의 기능과 스타일을 확장한 파이썬 시각화 도구

```python
import seaborn as sns
```

## seaborn import

```python
titanic = sns.load_dataset('titanic')
#sea born내부에 기본으로 제공되는 데이터
display(titanic.head())
display(titanic.info())
```

## seaborn 활용해보기 1
> 회귀선이 있는 산점도를 그려보자
- reglpot()함수 이용
- 서로 다른 2개의 연속 변수 사이의 산점도를 그리고 선형회귀분석에 의한 회귀선을 함께 나타낸다.
- 회귀선을 보이거나 안보이게 설정가능(fit_reg 옵션)

```python
import matplotlib.pyplot as plt
```

```python
sns.set_style('darkgrid')
```

- style의 종류 : darkgrid, whitegrid, dark, white, ticks

```python


fig= plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
# x축 변수<- 나이(age) y축 변수 <- 운임(fare)
sns.regplot(x='age', #x축 활용될 열이름
            y='fare', #y축 활용될 열이름
            data=titanic, #이용할 데이터
            ax=ax1) #axe객체 1번째 그래프
sns.regplot(x='age', #x축 활용될 열이름
            y='fare', #y축 활용될 열이름
            data=titanic, #이용할 데이터
            ax=ax2,#axe객체 2번째 그래프
            fit_reg=False) #회귀선 표시여부

plt.show()
```

## seaborn 활용해보기 2
> 히스토그램/커널밀도 그래프를 그려보자
- displot 함수 이용
- hist와 ked옵션으로 커널과 히스토그램여부를 조정
- 커널밀도 그래프 : 그래프와 x축사이의 면적이 1이 되도록 그리는 밀도 분포 그래프

```python
sns.set_style('whitegrid')

fig = plt.figure(figsize=(15,5))

ax1= fig.add_subplot(1,3,1)
ax2= fig.add_subplot(1,3,2)
ax3= fig.add_subplot(1,3,3)

sns.distplot(titanic['fare'], ax=ax1)
sns.distplot(titanic['fare'], ax=ax2, hist=False,)
sns.distplot(titanic['fare'], ax=ax3, kde=False)

ax1.set_title('titanic fare - hist/ked')
ax2.set_title('titanic fare - ked')
ax3.set_title('titanic fare - hist')

plt.show()
```

## seaborn 활용해보기 3
>히트맵 그래프를 그려보자
- heatmap 함수
- 2개의 범주형 변수를 각각 x,y축에 놓고 데이터를 메트릭스 형태로 분류
- 한 변수를 행 인덱스로, 다른변수를 열이름으로 설정
- aggfunc='size' 옵션은 데이터 값의 크기를 기준으로 집계한다는 뜻

```python
sns.set_style('whitegrid')

table= titanic.pivot_table(index=['sex'], columns=['class'], aggfunc='size')
table #피봇테이블은 추후설명.
```

```python
sns.heatmap(table, #데이터 프레임
           annot=True, #데이터 값 표시 여부
           fmt='d', #정수형 포맷
           cmap='YlGnBu', #컬러맵
           linewidth=5, # 구분선
           cbar=False)#컬러바 표시 여부
plt.show()
```

## seaborn 활용해보기 4
>범주형 데이터의 산점도
- stripplot() 함수 : 데이터 분산 미고려
- swarmplot() 함수 : 데이터 분산 고려(중복X)

```python
sns.set_style('whitegrid')

fig = plt.figure(figsize=(15,5))
ax1= fig.add_subplot(1,2,1)
ax2= fig.add_subplot(1,2,2)

sns.stripplot(x='class',
             y='age',
             data=titanic,
             ax=ax1)

sns.swarmplot(x='class',
             y='age',
             data=titanic,
             ax=ax2)

ax1.set_title('Stripplot')
ax2.set_title('Swarmplot')

plt.show()
```

## seaborn 활용해보기 6
>여러가지 막대 그래프
- barplot() 함수

```python
sns.set_style('whitegrid')

fig = plt.figure(figsize=(15,5))
ax1= fig.add_subplot(1,3,1)
ax2= fig.add_subplot(1,3,2)
ax3= fig.add_subplot(1,3,3)

# x축, y축에 변수를 할당해서 막대 그래프
sns.barplot(x='sex', y='survived', data=titanic, ax=ax1)

# x축, y축에 변수를 할당해서 막대 그래프, hue 옵션을 통해 class별 막대바 그래프
sns.barplot(x='sex', y='survived', data=titanic, ax=ax2, hue='class')

# x축, y축에 변수를 할당해서 막대 그래프 hue 옵션을 통해 class별 막대바 그래프 dodge=False를통해 누적그래프
sns.barplot(x='sex', y='survived', data=titanic, ax=ax3, hue='class', dodge=False)

plt.show()
```

## seaborn 활용해보기 7
> 기본값 빈도 그래프
- countplot() 함수

```python
sns.set_style('whitegrid')

fig = plt.figure(figsize=(15,5))
ax1= fig.add_subplot(1,3,1)
ax2= fig.add_subplot(1,3,2)
ax3= fig.add_subplot(1,3,3)

#Paletee 옵션은 colorpool을 선택하는것
sns.countplot(x='class', palette='Set1', data=titanic, ax=ax1)
sns.countplot(x='class', palette='Set2', data=titanic, ax=ax2, hue='who')
sns.countplot(x='class', palette='Set3', data=titanic, ax=ax3, hue='who', dodge=False)

plt.show()
```

## seaborn 활용해보기 8
> 박스 플롯
- boxplot()함수

```python
fig = plt.figure(figsize=(15,5))
ax1= fig.add_subplot(2,1,1)
ax2= fig.add_subplot(2,1,2)

sns.boxplot(x='alive', y='age', data=titanic, ax=ax1)
sns.boxplot(x='alive', y='age', data=titanic, ax=ax2, hue='sex')

plt.show()
```

## seaborn 활용해보기 9
> 바이올린 플롯
- violinplot()함수

```python
fig = plt.figure(figsize=(15,5))
ax1= fig.add_subplot(1,2,1)
ax2= fig.add_subplot(1,2,2)

sns.violinplot(x='alive', y='age', data=titanic, ax=ax1)
sns.violinplot(x='alive', y='age', data=titanic, ax=ax2, hue='sex')

plt.show()
```

## seaborn 활용해보기 10
> 산점도그래프.(joint)
- jointplot() 함수

```python
ax1=sns.jointplot(x='fare', y='age', data=titanic)
ax2=sns.jointplot(x='fare', y='age', data=titanic, kind='reg')
ax3=sns.jointplot(x='fare', y='age', data=titanic, kind='hex')
ax4=sns.jointplot(x='fare', y='age', data=titanic, kind='kde')

plt.show()
```

## seaborn 활용해보기 11
> FacetGrid를 사용하여 자동으로 분할해보기
- FacetGrid() 함수
- who조건이 (man, woman, child), survived조건(0,1)
- who조건이 col로, survived조건이 row줄로 나타나도록 그래프를 그려본다.

```python
g= sns.FacetGrid(data=titanic, col='who', row='survived')

g = g.map(plt.hist, 'age')

plt.show()
```
