# folium 라이브러리 사용하기
- folium의 라이브러리인 map()을 이요하면 간단한 지도 객체를 만들 수 있다.
- 화면은 줌과 화면이동 모두 가능하다.
- 단, folium은 웹기반 지도를 만들어 주기 때문에 ide에서 볼 수가 없다.
- 지도를 보기 위해 save()메소드를 이용하여 html로 저장을 하고, 열어서확인한다.

## 서울 지도 만들기

```python
seoul_map=folium.Map(location=[37.55, 126.98], zoom_start=12)
seoul_map
```

----------------
### tiles옵션 지정해보기

```python
seoul_map1=folium.Map(location=[37.55, 126.98],
                      zoom_start=12,
                     tiles='Stamen Terrain')
seoul_map1
```

```python
seoul_map2=folium.Map(location=[37.55, 126.98],
                      zoom_start=12,
                     tiles='Stamen Toner')
seoul_map2
```

## 마커 찍어보기

```python
import shutil
# shutil.move('C:/Users/i7B/Downloads/서울지역 대학교 위치.xlsx','서울지역 대학교 위치.xlsx')
```

```python
import pandas as pd
df=pd.read_excel('서울지역 대학교 위치.xlsx')
df.head()
```

```python
df.iloc[0:,0].head()
```

### 기본 Marker찍어보기

```python
for name, lat, lng in zip(df.iloc[0:,0], df.위도, df.경도):
    folium.Marker([lat,lng], popup=u'고').add_to(seoul_map)
```

```python
seoul_map
```

### 기본 CircleMarker찍어보기

```python
for name, lat, lng in zip(df.iloc[0:,0], df.위도, df.경도):
    folium.CircleMarker([lat,lng],
                        popup=name,
                        radium=10,
                        color='brown',
                        fill=True,
                       fill_color='coral',
                       fill_opacity=0.7).add_to(seoul_map)
seoul_map
```

```python
import json
df=pd.read_excel('경기도인구데이터.xlsx',index_col='구분')
df.head()
```

```python
df.columns = df.columns.map(str)
print(df.columns)
```

```python
g_map = folium.Map(location=[37.5502, 126.982],
                   tiles='Stamen Terrain',
                   zoom_start=9)
g_map
```

```python
# shutil.move('C:/Users/i7B/Downloads/경기도행정구역경계.json', '경기도행정구역경계.json')
```

```python
try:
    geo_data=json.load(open('경기도행정구역경계.json', encoding="utf-8"))
except:
    geo_data=json.load(open('경기도행정구역경계.json', encoding="utf-8-sig"))
    
year='2007'

folium.Choropleth(geo_data=geo_data,
                 data=df[year],
                 columns=[df.index, df[year]],
                 fill_color='YlOrRd',
                 fill_opacity=0.7,
                 line_opacity=0.3,
                 threshold_scale=[10000, 100000, 300000, 500000, 700000],
                 key_on='feature.properties.name'
                 ).add_to(g_map)

g_map
```
