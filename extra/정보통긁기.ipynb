```python
import requests
from bs4 import BeautifulSoup
```

```python
data={
    'bbs_id': 'T2014-0844-04-785706',
    'page': '1',
    'page_size': '500',
    'notice_yn': 'N',
    'search_period': 'total',
    'search_key': 'title_contents',
    'search_value': '부산광역시 동구',
    'kbs_board_auth': '0'
    
}
```

```python
rep=requests.get('https://cfpbbsapi.kbs.co.kr/board/v1/list',params=data)
```

```python
page=BeautifulSoup(rep.content,'html.parser')
```

```python
page
```

```python
import json
```

```python
asdf=json.loads(page.text)
```

```python
asdf
```

```python
how_many=asdf['total_count']
```

```python
target_ids=[]
```

```python
for a in asdf['data'][4:]:
    target_ids.append(a['target_url'][-6:])
```

```python
target_ids
```

```python
import re
```

```python
list_texts=[]
```

```python
for idd in target_ids:
    t_data={
    'bbs_id': 'T2014-0844-04-785706',
    'id': idd,
    'post_no': '',
    'search_period':'', 
    'search_key': '',
    'search_value': ''
    }
    t_rep=requests.get('https://cfpbbsapi.kbs.co.kr/board/v1/read_post',params=t_data)
    efgh=json.loads(t_rep.text)
    list_texts.append(efgh['post']['post_contents'])
```

```python
list_texts
```

```python
for x in list_texts:
    y=BeautifulSoup(x,'html.parser')
    for z in y.find_all('p'):
        if z.get_text()=='\xa0':
            z.extract()
    for z in range(len(y.find_all('p'))):
        if '부산' in y.find_all('p')[z].get_text():
            try :
                print(y.find_all('p')[z-2].get_text())
                print(y.find_all('p')[z-1].get_text())
                print(y.find_all('p')[z].get_text())
                print(y.find_all('p')[z+1].get_text())
            except:
                print(y.find_all('p')[z].get_text(),'문제')
            print('==================')
```
