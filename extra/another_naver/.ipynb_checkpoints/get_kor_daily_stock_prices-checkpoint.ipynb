```python
import requests
from bs4 import BeautifulSoup
import time
import random
import pickle
import sqlite3
```

- DB만들기

```python
conn = sqlite3.connect("stocks_from_naver.db")
cur = conn.cursor()
```

* 주요 테이블 생성

```python
cur.execute('drop table kor_daily_stock_prices')
conn.commit()
```

```python
cur.execute("create table kor_daily_stock_prices(code int, company char(30), date char(20),price int)")
```

```python
with open('../naver_stock/step_1_done.pickle', 'rb') as handle:
    test_dict=pickle.load(handle)
soup_dict=test_dict.copy()
```

```python
soup_dict
```

```python
for row in cur.fetchall():
    print(row)
```

```python
for x in soup_dict:
    time.sleep(random.randint(1000,2000)*0.001)
    try:
        for i in range(1,19):
            page=requests.get('https://finance.naver.com/item/sise_day.nhn?code={}&page={}'.format(soup_dict[x][1],i))
            html=page.text
            page_soup=BeautifulSoup(html,'html.parser')
            a=page_soup.findAll('tr')
            for b in [2,3,4,5,6,10,11,12,13,14]:
                c=a[b].findAll('td')
                cur.execute("insert into kor_daily_stock_prices values({},'{}','{}',{})".format(soup_dict[x][1],x,c[0].get_text().strip(),c[1].get_text().strip().replace(',','')))
                print(c[0].get_text().strip())
    except:
        print(x,'fali')
    print(x)
    conn.commit()
```

```python
cur.execute("select * from kor_daily_stock_prices where date like '2019.12.30'")
```
