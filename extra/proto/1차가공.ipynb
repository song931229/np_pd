```python
import pandas as pd
```

```python
df=pd.read_excel('naver_stock.xlsx',index_col=0)
```

```python
df
```

```python

for si in  df.iloc[:,2:10]:
    try:
        df[si]=df[si].str.replace('[^0-9.]','')
    except:
        pass
for si in  df.iloc[:,11:]:
    try:
        df[si]=df[si].str.replace('[^0-9.]','')
    except:
        pass
```

```python
for si in  df.iloc[:,2:]:
    try:
        df[si]=df[si].astype('float')
    except:
        pass
```

```python
df.info()
```

```python
df['투자의견'].fillna(value='0중립',inplace=True)
list(df['투자의견'])
```

```python
df.fillna(value=0,inplace=True)
df.head()
```

```python
df['투자의견']
```

```python
df['투자의견']=df['투자의견'].str.replace('매수','/매수',regex=False)
df['투자의견']=df['투자의견'].str.replace('매도','/매도',regex=False)
df['투자의견']=df['투자의견'].str.replace('중립','/중립',regex=False)
```

```python
df.head(2)
```

```python
df['투자의견수치'] = df.투자의견.str.split('/').str[0]
```

```python
df['투자의견방향'] = df.투자의견.str.split('/').str[1]
```

```python
df.to_excel('mid_naver_stock.xlsx')
```
