# 시총순위에 따른 URL List만들기.

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pickle
import re
import time
import random
urls=["https://finance.naver.com/sise/sise_market_sum.nhn?&page="+str(x) for x in range(1,11)]
```

## 각 URL에따른 종목코드 따오기

```python
soup_dict={}

for j,url in enumerate(urls):

    page = urlopen(url)

    page_html = page.read()

    page_soup = BeautifulSoup(page_html, 'html.parser')
    
    atags=page_soup.find_all('a','tltle')
    
    for i,atag in enumerate(atags):
        title=atag.get_text()
        if title not in soup_dict:
            soup_dict[title]=[(j*50)+i+1,atag.get_attribute_list('href')[0][-6:]]
        else:
            soup_dict[title+"2"]=[(j*50)+i+1,atag.get_attribute_list('href')[0][-6:]]
    time.sleep(random.randint(500,2500)*0.001)
```

# 종목코드 기반 접속 후 ID,ENCPARAM 따오기

### id와 enc가져오기

```python
for x in soup_dict:
    zt=time.time()
    time.sleep(random.randint(2000,3500)*0.001)

    inner_page=urlopen(soup_dict[x][2])

    inner_page_soup = BeautifulSoup(inner_page, 'html.parser')
    
    try:
        com_id=inner_page_soup.select('#wrapper>div')[13]['id']
    except:
        print(x)
    #어펜드로보내기.
    soup_dict[x].append(com_id)
    
    
    #enc파람구하기
    target_script=inner_page_soup.select('script')[6]
    
    target_text=target_script.get_text().strip()
    
    target_index=target_text.find('encparam')
    
    com_encparam=target_text[target_index+11:target_index+43]
    
    soup_dict[x].append(com_encparam)
    st=time.time()
    time.sleep(random.randint(100,200)*0.001)
    url='https://navercomp.wisereport.co.kr/v2/company/ajax/cF1001'\
    '.aspx?cmp_cd={}&fin_typ=0&freq_typ=A&encparam={}&id={}'.format(soup_dict[x][1],soup_dict[x][4],soup_dict[x][3])
    print(url)
    page = urlopen(url)

    page_html = page.read()

    page_soup = BeautifulSoup(page_html, 'html.parser')
    #인덱스구하기

    #2019/12
    b=page_soup.select('table:nth-child(2) > tbody > tr > td:nth-child(7)')
    #2020/03
    c=page_soup.select('table:nth-child(2) > tbody > tr > td:nth-child(8)')
    #2020/06
    d=page_soup.select('table:nth-child(2) > tbody > tr > td:nth-child(9)')
    
    for bb in b:
        soup_dict[x].append(bb.get_text().strip())
    for cc in c:
        soup_dict[x].append(cc.get_text().strip())
    for dd in d:
        soup_dict[x].append(dd.get_text().strip())
    et=time.time()
    print(soup_dict[x])
    print(st-zt,et-st)
```

```python
drop_keys=[]
for com in soup_dict:
    if ')' in soup_dict[com][4]:
        drop_keys.append(com)
```
