```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
import time
import random
import pickle
import pandas as pd
```

```python
with open('filtered_rand_code_url.pickle', 'rb') as handle:
    soup_dict=pickle.load(handle)
```

```python
soup_dict
```

```python
for key in soup_dict:
    time.sleep(random.randint(500,2500)*0.001)
    url='https://navercomp.wisereport.co.kr/v2/company/ajax/cF1001'\
    '.aspx?cmp_cd={}&fin_typ=0&freq_typ=A&encparam={}&id={}'.format(soup_dict[key][1],soup_dict[key][4],soup_dict[key][3])
    
    
    page = urlopen(url)

    page_html = page.read()

    page_soup = BeautifulSoup(page_html, 'html.parser')
    #인덱스구하기

    #2019/12
    b=page_soup.select('table:nth-child(2) > tbody > tr > td:nth-child(7)')
    #2020/03
    c=page_soup.select('table:nth-child(2) > tbody > tr > td:nth-child(8)')
    #2020/06
    d=page_soup.select('table:nth-child(2) > tbody > tr > td:nth-child(9)')
    
    for bb in b:
        soup_dict[key].append(bb.get_text().strip())
    for cc in c:
        soup_dict[key].append(cc.get_text().strip())
    for dd in d:
        soup_dict[key].append(dd.get_text().strip())
    
    print(soup_dict[key])
```

```python
with open('new_naver_after_half.pickle', 'wb') as handle:
    pickle.dump(soup_dict, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
