```python
from selenium import webdriver
from bs4 import BeautifulSoup

url='https://kr.investing.com//equities/samsung-heavy-industries'

driver=webdriver.Chrome('C:/Chrome Driver/chromedriver')

driver.implicitly_wait(1)

driver.get(url)
```

# 셀레니움단.

```python
mini_table=driver.find_element_by_xpath('/html/body/div[5]/section/div[9]') 
mini_table_text = mini_table.text
```

```python
slices=mini_table_text.split('\n')
```
