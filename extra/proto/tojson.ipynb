```python
import json
import requests

news_id_li = [] #뉴스 아이디 리스트를 만들었다
get_news_list=[]
def get_news_id():
    key_word = str(input('key_word : ')) # str = 문자형으로 입력받아 key_word 함수에 넣는다.
    key_word = '"' + key_word + '"' # key_word 사이에 "" 추가하기
    print('ex) 2020-02-16') #날짜 예시 출력해 보여주기
    start_date = str(input('start date : ')) # 시작 날짜 문자형으로 받기
    start_date = '"' + start_date + '"' # 시작 날짜에도 "" 추가해주기
    end_date = str(input('end date : ')) # 끝 날짜 문자형으로 받기
    end_date = '"' + end_date + '"' # 끝 날짜에도 ""추가해주기
    url = "https://www.bigkinds.or.kr/api/news/search.do" #URL (빅카인즈)
    params = '''{"indexName": "news", "searchKey": %s, "searchKeys": [{}], "byLine": "", "searchFilterType": "1", "searchScopeType": "1", "mainTodayPersonYn": "", "startDate": %s, "endDate": %s, "newsIds": [], "categoryCodes": [], "incidentCodes": [], "networkNodeType": "", "topicOrigin": "", "startNo": "1", "resultNumber": "1"}''' %(key_word, start_date, end_date) #첫페이지?
    req = requests.post(url, data = json.dumps(eval(params))) # 형식?
    sol = req.json() #정보 요청
    t_count =  '"' + str(sol['totalCount']) + '"' # 끌어올 크기
    go_params = '''{"indexName": "news", "searchKey": %s, "searchKeys": [{}], "byLine": "", "searchFilterType": "1", "searchScopeType": "1", "mainTodayPersonYn": "", "startDate": %s, "endDate": %s, "newsIds": [], "categoryCodes": [], "incidentCodes": [], "networkNodeType": "", "topicOrigin": "", "startNo": "1", "resultNumber": %s}''' %(key_word, start_date, end_date, t_count) #다음 페이지
    go_req = requests.post(url, data = json.dumps(eval(go_params))) #프린트 할 형식 (디코딩 했던 파이썬 데이터를 json으로 바꿈)
    go_sol = go_req.json()

# json은 {"name":"value"}의 형식으로 구성, 각각은 콤마로 구분분
    for tag in go_sol['resultList']:
        news_id = tag['NEWS_ID']
        news_id_li.append(news_id) #데이터 통합하기

def crawl(the_list): #
    global get_news_list
    for n_id in the_list: # n_in(함수)에 the_list에 있는것을  넣어
        cl_url = "https://www.bigkinds.or.kr/news/detailView.do?docId=%s&returnCnt=1&sectionDiv=1000" %n_id #?
        cl_req = requests.get(cl_url) #정보 요청
        json_cl = cl_req.json()
        get_news_list.append(json_cl)
        print(json_cl['detail']['CONTENT'])
        print('---------------------------------------------------')
get_news_id()
crawl(news_id_li)
```
