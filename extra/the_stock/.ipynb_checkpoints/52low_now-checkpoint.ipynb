```python
import time
import random
```

```python
from selenium import webdriver
from bs4 import BeautifulSoup

url='https://kr.investing.com/equities/south-korea'

driver=webdriver.Chrome('C:/Chrome Driver/chromedriver')

driver.implicitly_wait(1)

driver.get(url)
```

```python
#select option클릭
driver.implicitly_wait(5)
sel_box = driver.find_element_by_xpath('/html/body/div[5]/section/div[6]/select/option[11]') 
sel_box.click()
```

```python
import pandas as pd
```

# 가져오기

```python
# 수프로 가져오기
html=driver.page_source

a=BeautifulSoup(html,'html.parser')
```

## 태그가공

### 수프로 태그가공

```python
b=a.find_all("td", class_="bold left noWrap elp plusIconTd")
```

```python
ds=[]
for c in b:
    d=c.find('a')
    ds.append(d)
```

```python
fs={}
for e in ds:
    f=e.attrs
    if f['title'] not in fs:
        fs[f['title']]=[f['href']]
    else:
        fs[f['title']+'2']=[f['href']]
```

```python
fs
```

```python
soup_atags=[]
for ff in fs:
    fs[ff][0]='https://kr.investing.com/'+fs[ff][0]
```

```python
soup_atags=fs
```

### 셀레니움으로 태그가공

```python
# 셀레니움으로 가져오기
driver.implicitly_wait(3)
trs=[str(x) for x in range(1,201)]
atags={}
```

```python
for tr in trs:
    aa=driver.find_element_by_xpath('/html/body/div[5]/section/div[8]/div/table/tbody/tr['+tr+']/td[2]/a')
    if aa.text not in atags:
        atags[aa.text]=[aa.get_attribute('href')]
    else:
        atags[aa.text+'2']=[aa.get_attribute('href')]
```

```python
atags
```

```python
len(atags)
```

```python
sel_atags=atags
```

## 테이블가공하기

### selenium단에서 테이블가공

```python
table=driver.find_element_by_xpath('/html/body/div[5]/section/div[8]/div/table') 
table_text = table.text
```

```python
table_data1=table_text.split('\n')
table_data1
```

```python
for x in range(len(table_data1)):
    table_data1[x]=table_data1[x].strip()
```

```python
table_data1
```

```python
print(len(table_data1))
```

```python
table_data2=[]
for x in range(1,len(table_data1)):
    table_data2.append(table_data1[x].split(' '))
```

```python
len(table_data2)
```

```python
for x in table_data2:
    if len(x)!=8 and len(x)!=9:
        print(x)
```

```python
mytable1_data={}
```

```python
for x in table_data2:
    if len(x)==8:
        if x[0] not in mytable1_data:
            mytable1_data[x[0]]=[x[i] for i in range(1,8)]
        else :
            mytable1_data[x[0]+'2']=[x[i] for i in range(1,8)]
    elif len(x)==9:
        if (x[0]+' '+x[1]) not in mytable1_data:
            mytable1_data[x[0]+' '+x[1]]=[x[i] for i in range(2,9)]
        else:
            print(x[0]+' '+x[1],2)
```

```python
len(mytable1_data)
```

```python
my_table1=pd.DataFrame(mytable1_data)
my_table1=my_table1.T
my_table1
```

```python
my_table1.columns=['현재가','고가','저가','변동','변동%','거래량','시간']
```

```python
my_table1.head()
```

### soup단에서 테이블가공

```python
stable_data=a.find("table", class_="genTbl closedTbl crossRatesTbl elpTbl elp25")
```

```python
tds=stable_data.find_all('td')
```

```python
for td in tds:
    print(td.text.strip(),'<',sep='')
```

```python
len(tds)
```

```python
tds_text=[]
for td in tds:
    tds_text.append(td.text.strip())
```

```python
dictsbysoup={}
```

```python
for x in range(0,2000,10):
    if tds_text[x+1] not in dictsbysoup:
        dictsbysoup[tds_text[x+1]]=[tds_text[x+y] for y in range(2,9)]
    else:
        dictsbysoup[tds_text[x+1]+'2']=[tds_text[x+y] for y in range(2,9)]
```

```python
dictsbysoup
```

```python
len(dictsbysoup)
```

```python
my_table2=pd.DataFrame(dictsbysoup)
```

```python
my_table2=my_table2.T
my_table2.head()
```

```python
my_table2.columns=['현재가','고가','저가','변동','변동%','거래량','시간']
```

```python
my_table2.head()
```

```python
for key in sel_atags:
    time.sleep(random.randint(1,2))
    driver.get(sel_atags[key][0])
    time.sleep(random.randint(1,3))
    mini_table=driver.find_element_by_xpath('/html/body/div[5]/section/div[9]') 
    mini_table_text = mini_table.text
    slices=mini_table_text.split('\n')
    for i in range(1,len(slices),2):
        sel_atags[key].append(slices[i])
```
