```python
import pandas as pd
```

```python
df=pd.read_excel('data.xlsx')
```

```python
df.tail()
```

```python
df=df.rename(columns={'Unnamed: 0':'종목'})
```

```python
df.head()
```

```python
df.columns
```

```python
myneeds=df.loc[:,['종목','총 시가','현재가','52주 변동폭']]
```

```python
myneeds.head()
```

```python
for_add=myneeds.iloc[:,3].str.split(expand=True)
```

```python
for_add
```

```python
for_add.drop(1,axis=1,inplace=True)
```

```python
for_add.columns=['저가','고가']
```

```python
for_add.head(2)
```

```python
for_add['고가']=for_add['고가'].str.replace(pat=r'[^A-Za-z0-9]', repl=r'', regex=True)
```

```python
for_add['저가']=for_add['저가'].str.replace(pat=r'[^A-Za-z0-9]', repl=r'', regex=True)
```

```python
for_add.astype({'고가':'float','저가':'float'})
```

```python
all_needs=pd.merge(myneeds, for_add, right_index=True,left_index=True)
```

```python
all_needs['총 시가']=all_needs['총 시가'].str.replace(pat=r'[^0-9.]', repl=r'', regex=True)
```

```python
all_needs.rename(columns={'총 시가':'총시가(B)'})
```

```python
all_needs['현재가']=all_needs['현재가'].str.replace(pat=r'[^A-Za-z0-9]', repl=r'', regex=True)
```

```python
all_needs=all_needs.astype({'현재가':'float','저가':'float','고가':'float','총시가(B)':'float'})
```

```python
all_needs['손실']=all_needs['고가']-all_needs['저가']
```

```python
all_needs['회복']=all_needs['현재가']-all_needs['저가']
```

```python
all_needs['손실율%']=100-(all_needs['저가']/all_needs['고가']*100)
```

```python
all_needs.head()
```

```python
all_needs['회복율%']=all_needs['회복']/all_needs['손실']*100
```

```python
all_needs['현재/고가%']=all_needs['현재가']/all_needs['고가']*100
all_needs['저가/고가%']=(all_needs['저가']/all_needs['고가'])*100
```

```python
all_needs.head()
```

```python
all_needs.to_excel('52주 현황.xlsx')
```
