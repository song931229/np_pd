```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import re
%matplotlib inline
plt.rc("font", family="Malgun Gothic")
plt.rc("axes", unicode_minus=False)
%config InlineBackend.figure_format = 'retina'
```

```python
df=pd.read_csv('monthly_accident.txt', '\t')
```

```python
df.head(4)
```

```python
df=df.drop([0,1,2])
```

```python
df.head(1)
```

```python
df=df.drop('기간',axis=1)
```

```python
df.head(1)
```

```python
for x in range(8):
    try:
        df.iloc[:,x]=df.iloc[:,x].str.replace('-','0')
        df.iloc[:,x]=df.iloc[:,x].str.replace(',','')
        df.iloc[:,x]=df.iloc[:,x].astype('float')
    except:
        pass
```

```python
pivot=df.pivot(index='자치구별',columns='구분')
```

```python
mon=int(input("몇월?"))

fig=plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(1, 1, 1)
ax1.bar([x*3 for x in range(25)], pivot[str(mon)+'월','발생건수'],tick_label=pivot.index)
ax1.bar([x*3+1 for x in range(25)], pivot[str(mon)+'월','부상자수'])
ax2=ax1.twinx()
ax2.plot([x*3+0.5 for x in range(25)], pivot[str(mon)+'월','사망자수'],color='red')

ax1.set_xticklabels(ax1.get_xticklabels(),rotation=45)

plt.show()
```
