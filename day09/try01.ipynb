```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import re

df = pd.read_csv('car_crush_by_License.txt', sep="\t")

df.drop([0, 1, 2], axis=0, inplace=True)
df.drop(['기간', '합계'], axis=1, inplace=True)
df.rename(columns={df.columns[0]: '구별'}, inplace=True)

df_occ = df.loc[df.구분 == '발생건수', :].set_index('구별').drop('구분', axis=1)
df_occ = df_occ.applymap(lambda x: re.sub(",", "", x)).astype('int')

fig = plt.figure(figsize=(16, 10))

def create_x(t, w, n, d):
    return [t * x + w * n for x in range(d)]


value_a_x = create_x(5, 0.8, 1, 25)
value_b_x = create_x(5, 0.8, 2, 25)
value_c_x = create_x(5, 0.8, 3, 25)
value_d_x = create_x(5, 0.8, 4, 25)
value_e_x = create_x(5, 0.8, 5, 25)

ax = plt.subplot()
ax.bar(value_a_x, df_occ['5년 미만'], label='5년미만')
ax.bar(value_b_x, df_occ['10년 미만'], label='10년미만')
ax.bar(value_c_x, df_occ['15년 미만'], label='15년미만')
ax.bar(value_d_x, df_occ['15년 이상'], label='15년이상')
ax.bar(value_e_x, df_occ['기타불명'], label='기타불명')

middle_x = [(a + b + c + d + e) / 5 for (a, b, c, d, e) in zip(value_a_x, value_b_x, value_c_x, value_d_x, value_e_x)]

ax.set_xticks(middle_x)
ax.set_xticklabels(df_occ.index)

plt.legend()

plt.show()

'''
df_death = df.loc[df.구분=='사망자수', :].set_index('구별').drop('구분', axis=1)
df_injuly = df.loc[df.구분=='부상자수', :].set_index('구별').drop('구분', axis=1)
'''
```
