# 데이터 프레임+데이터 프레임

```python
from IPython.display import display_html
def display_side(*args):
    html_str=''
    for df in args:
        html_str+=df.to_html()
    display_html(html_str.replace('table','table style="display:inline"'),raw=True)
```

```python
import pandas as pd

df1 = pd.DataFrame({'a': ['a0', 'a1', 'a2', 'a3'],
                    'b': ['b0', 'b1', 'b2', 'b3'],
                    'c': ['c0', 'c1', 'c2', 'c3']},
                    index=[0, 1, 2, 3])

df2 = pd.DataFrame({'a': ['a2', 'a3', 'a4', 'a5'],
                    'b': ['b2', 'b3', 'b4', 'b5'],
                    'c': ['c2', 'c3', 'c4', 'c5'],
                    'd': ['d2', 'd3', 'd4', 'd5']},
                    index=[2, 3, 4, 5])
display_side(df1,df2)
```

## concat의 사용

```python
df3=pd.concat([df1,df2],axis=0)
df4=pd.concat([df1,df2],axis=0, join='outer') #합집합 - defalt
df5=pd.concat([df1,df2],axis=0, join='inner') #교집합
```

### Join옵션
- inner와 outer를 통해 열의 교집합과 합집합을 구분할수있음

```python
display(df3)
display_side(df4,df5)
```

### ignor_index 옵션

- 기존의 인덱스를 무시하고 새로 설정해서 만듦

```python
df6=pd.concat([df1,df2],axis=0,ignore_index=True)
```

# 데이타프레임+시리즈

```python
sr1 = pd.Series(['e'+str(x) for x in range(4)], name='e')
sr2 = pd.Series(['f'+str(x) for x in range(3)], name='f', index=[3,4,5])
sr3 = pd.Series(['g'+str(x) for x in range(4)], name='g')
```

## concat의 활용

```python
result1=pd.concat([df1,sr1],axis=1)
result2=pd.concat([df1,sr2],axis=1)
result3=pd.concat([df1,sr3],axis=1)
display_side(result1,result2,result3)
```

```python
result4=pd.concat([sr1,sr3],axis=1)
result5=pd.concat([sr2,sr3],axis=0)
display(result4)
result5
```
