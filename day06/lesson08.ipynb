# 데이터의 분할

```python
import pandas as pd
import seaborn as sns
```

```python
titanic=sns.load_dataset('titanic')
```

```python
df=titanic.loc[:,['age','sex','class','fare','survived']]
```

```python
df.head()
```

```python
len(df)
```

## Group by

```python
grouped=df.groupby(['class'])
grouped
```

```python
for key,group in grouped:
    print("* key :", key)
    print("* number :", len(group))
    display(group.head())
    print()
```

### 그룹별로 연산 메소드 적용하기

```python
average=grouped.mean()
average
```

### 그룹내의 특정값 꺼내기

- key값을 활용하여 그룹꺼내기

```python
group3 = grouped.get_group('Third')
```

```python
group3.head()
```

### Group by 2
- 여러열을 기준으로도 나눌수 있다.

```python
grouped2=df.groupby(['class','sex'])
for key,group in grouped2:
    print("* key :", key)
    print("* number :", len(group))
    display(group.head())
    print()
```

```python
grouped2.mean()
```
