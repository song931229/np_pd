## 열과 행 추가하기
- df[열이름] = data
- df.loc[행이름] = data

### 열추가

```python
import numpy as np
import pandas as pd

exam_data = {'이름':['학생1','학생2','학생3'],'수학': [90, 80, 70], '영어': [98, 89, 95],
             '음악': [85, 95, 100], '체육': [100, 95, 95]}
df = pd.DataFrame(exam_data)
df
```

```python
df['국어']=80
df
```

```python
df['컴퓨터']=[100,70,80]
df
```

### 행추가하기
- df.loc[행이름] = data

```python
df.loc[3]=0
df
```

```python
df.loc[4]=['학생5',90,80,70,60,50,40]
df
```

```python
df.loc[5]=df.loc[3]
df
```
