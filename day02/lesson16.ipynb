## 행렬을 정렬하기

```python
import pandas as pd
import numpy as np

dict_data = {'a': [1, 2, 3], 'b': [4, 5, 6], 'c': [7, 8, 9],\
             'd': [10, 11, 12], 'e': [13, 14, 15]}
df = pd.DataFrame(dict_data, index=['r0', 'r1', 'r2'])
df
```

```python
ndf = df.sort_index(ascending=False)
ndf
```

```python
ndf2= ndf.sort_index()
ndf2
```

```python
exam_data = {'이름':['유재석','강호동','홍길동'],'수학': [90, 80, 70], '영어': [98, 89, 95],
             '음악': [85, 95, 100], '체육': [100, 95, 95]}
df = pd.DataFrame(exam_data)
df
```

```python
df.set_index('이름',inplace=True)
df
```

```python
ndf=df.sort_index(ascending=False)
ndf
```

```python
ndf2=df.sort_values(by='영어',ascending=False)
ndf2
```

```python
ndf2=df.sort_values(by='홍길동',ascending=False,axis=1)
ndf2
```
