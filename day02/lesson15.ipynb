## 특정 열을 행 인덴스로 설정
- set_index([열이름] or 열이름)
- 결과물을 반환한다.
- set_index사용시 기존이 행인덱스는 소멸

```python
import numpy as np
import pandas as pd

exam_data = {'이름':['학생1','학생2','학생3'],'수학': [90, 80, 70], '영어': [98, 89, 95],
             '음악': [85, 95, 100], '체육': [100, 95, 95]}
df = pd.DataFrame(exam_data)
df
```

```python
ndf=df.set_index(['이름'])
ndf
```

```python
ndf=df.set_index(['음악'])
ndf
```

```python
dict_data={'a':[1,2,3],'b':[4,5,6],'c':[7,8,9],'d':[10,11,12],'e':[13,14,15]}
df= pd.DataFrame(dict_data, index=['r0','r1','r2'])
df
```

```python
new_index=['r0','r1','r2','r3','r4']
ndf=df.reindex(new_index)
ndf
```

```python
ndf2=df.reindex(new_index, fill_value=0)
ndf2
```

```python
ndf3=df.reset_index() # index를 정수형으로 초기화
ndf3
```
